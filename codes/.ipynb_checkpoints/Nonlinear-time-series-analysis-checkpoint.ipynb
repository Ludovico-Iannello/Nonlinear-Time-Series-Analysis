# Nonlinear Time series analysis

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.signal import argrelextrema
from time_series_analysis_functions import *
from scipy.fft import fft, fftfreq
from scipy.optimize import curve_fit
```

```python
%matplotlib inline
```

```python
#lenght of the time series (we consider xs as the time series observed)
N=50000
dt=0.01
xs, ys, zs = Rossler_attractor(dt=dt,N=N)
# embedding parameters
m=3
tau=148
M=len(xs)-(m-1)*tau
```

```python
fig = plt.figure(1)
ax = plt.axes(projection='3d')
ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel('x', fontsize=10)
ax.set_ylabel('y', fontsize=10)
ax.set_zlabel('z', fontsize=10)
plt.tick_params(labelsize=10)
ax.set_title('Rossler Attractor', fontsize=15)
#plt.savefig("Rossler_attractor.png",dpi=300)
plt.show()
```

## Phase reconstruction

```python
fig = plt.figure(2)
X_matr=embedding(xs,m,tau)
plt.errorbar(X_matr[:,0],X_matr[:,1],lw=0.5)
#plt.savefig("phase_reconstruction.png",dpi=300)
plt.show()
```

## Mutual information

```python
#mutual information
I,tau_arr=time_mi(xs)
tau_min=argrelextrema(I, np.less, order=20)
print('taumin',tau_min)

fig = plt.figure(3)
plt.errorbar(tau_arr,I)
plt.show()
```

## Correlation to determine best time delay

```python
#correlation
corr,tau_arr=corr_arr(xs)
tau_min=argrelextrema(corr, np.less, order=10)
print('taumin',tau_min)

fig = plt.figure(4)
plt.errorbar(tau_arr,corr)
plt.show()
```

```python
fnn_arr=np.zeros(9)
r_arr=np.zeros(9)
for r in range(1,10,1):
    X_fnn=fnn(3,xs,r,tau=69)
    fnn_arr[r-1]=X_fnn
    r_arr[r-1]=r
    
plt.figure(5)    
plt.errorbar(r_arr,fnn_arr)
plt.xlabel('r')
plt.ylabel(r'$X_{fnn}(r)$')
plt.show()    
```

```python
fft_result = np.fft.rfft(xs)
freq =np.linspace(0,1/(2*dt),len(fft_result))
plt.figure(6)
plt.plot(freq, np.abs(fft_result))
plt.title("Spettro di Frequenza (FFT)")
plt.xlabel('f [Hz]')
plt.ylabel("fft")
plt.xlim(0, 2)
plt.tight_layout()
main_freq=freq[np.argmax(fft_result)]


print('main freq',main_freq)
print('main period',1/main_freq)

plt.show()
```

```python
#embedding varing dimension m
'''
X=embedding(xs,3,tau)
d,t=mean_lyapunov_exp(X,max_l=1000,window=6)
X=embedding(xs,2,tau)
d2,t2=mean_lyapunov_exp(X,max_l=1000,window=6)
X=embedding(xs,1,tau)
d1,t1=mean_lyapunov_exp(X,max_l=1000,window=6)
'''
plt.figure(7)
plt.errorbar(dt*t1,d1,label='m=1')
plt.errorbar(dt*t2,d2,label='m=2')
plt.errorbar(dt*t,d,label='m=3')
plt.ylim(-5,2)
plt.xlabel('Tempo [s]')
plt.ylabel(r'$\langle log(d_j(t)) \rangle$')
plt.tight_layout()
plt.legend()
plt.show()
```

## Fitting largest lyapunov exponent

```python
def fit(t, c, lyap):
    return (c+lyap*t)

pars, covm =curve_fit(fit,dt*t2,d2, absolute_sigma=True)

print('best fit parameters',*pars)
print('errors',np.sqrt(covm.diagonal()))
```
